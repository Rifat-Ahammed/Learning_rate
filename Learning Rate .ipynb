```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the loss function and its derivative

def loss_function(w):
    
    return (w - 4)**2
```

```python
def derivative_loss_function(w):
    
    return 2 * (w - 4)
```

```python
# Intialize variable

initial_w = 0
learning_rate = [0.01, 0.1, 0.5, 1.0]
iterations = 10
```

```python
# Perform the gradient descent

paths = []

for lr in learning_rate:
    
    w = initial_w
    path = [w]
    
    for i in range(iterations):
        w = w - lr * derivative_loss_function(w)
        path.append(w)
        
    paths.append(path)
    
```

```python
print(paths)
```

```python
# plotting

plt.figure(figsize=(10, 6))

for i, lr in enumerate(learning_rate):
    #print(i)
    plt.plot(paths[i], label=f'LR={lr}')
    
plt.xlabel('Iteration')
plt.ylabel('Weight Value')
plt.title('Weight update paths for DIfferent learning Rates')
plt.legend()
plt.grid(True)
plt.show()
```
